```python
import pandas as pd
df = pd.read_csv('sample_data/citrus.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
# JST tidak bisa menghandle string, jadi ubah ke numerik
df.name[df.name == 'orange'] = 0
df.name[df.name == 'grapefruit'] = 1
```

```python
# Data masih dalam dataframe, harus diubah ke larik agar bisa diproses model
dataset = df.values
```

```python
dataset
```

```python
# Pisahkan atribut dan label, dari data tersebut kolom 5 terahir adalah atribut
X = dataset[:, 1:6]
# bilangan sebelum koma untuk memilih baris pada dataframe
# bilangan setelah koma untuk memilih kolom pada dataframe
```

```python
y = dataset[:, 0]
```

```python
# Normalisasi
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
# Pisahkan data latih dan data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scale, y, test_size=0.3)
```

```python
# Karena tadi kolom name diubah menjadi numerik yang mana adalah boolean, maka perlu diubah lagi menjadi float32
import numpy as np
y_train = y_train.astype(np.float32)
y_test = y_test.astype(np.float32)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=[5,]),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```
