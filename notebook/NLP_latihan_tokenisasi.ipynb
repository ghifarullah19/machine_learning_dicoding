```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# Membuat token sebanyak 15 kata dengan mengganti kata tidak mengerti dengan '-'
tokenizer = Tokenizer(num_words=15, oov_token='-')
```

```python
teks = [
    'Saya suka programming',
    'Programming sangat menyenangkan',
    'Machine Learning berbeda dengan pemrograman konvensional'
]
```

```python
tokenizer.fit_on_texts(teks)
```

```python
sequences = tokenizer.texts_to_sequences(teks)
```

```python
print(tokenizer.word_index)
```

```python
print(tokenizer.texts_to_sequences(['Saya suka programming']))
print(tokenizer.texts_to_sequences(['Saya suka belajar programming sejak SMP']))
```

```python
# Menambah padding di awal
from tensorflow.keras.preprocessing.sequence import pad_sequences
sequences_sama_panjang = pad_sequences(sequences)
```

```python
print(sequences_sama_panjang)
```

```python
# Menambah padding di akhir
sequences_sama_panjang = pad_sequences(
    sequences,
    padding='post',
    maxlen=5
)
```

```python
sequences_sama_panjang
```

```python
# Jika kata lebih dari maxlen maka ambil 5 kata awal saja
sequences_sama_panjang = pad_sequences(
    sequences,
    padding='post',
    maxlen=5,
    truncating='post'
)
```

```python
sequences_sama_panjang
```
